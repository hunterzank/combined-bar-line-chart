```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    'Month':['Jan','Feb','March','April','May', 'June']*2,
    'Year':[2021]*6 + [2022]*6,
    'Total Sales':[200,180,220,260,300,280,210,230,250,270,320,310],
    'Customer Count':[50,45,55,60,65,60,52,58,62,65,70,68]
}
df=pd.DataFrame(data)
fig, ax1=plt.subplots()
ax1.bar(df['Month'],df['Total Sales'], color = 'b', label='Total Sales')
ax2 = ax1.twinx()
ax2.plot(df['Month'],df['Customer Count'], color = 'g', label='Customer Count')
plt.title('Monthly Sales Data & Customer Count')
ax1.set_xlabel('Month')
ax1.set_ylabel('Total Sales', color = 'b')
ax2.set_ylabel('Customer Count', color = 'g')
ax1.legend(loc = 'upper left')
ax2.legend(loc = 'upper right')
plt.show
```
